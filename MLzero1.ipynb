# 坤和的机器学期零基础笔记1



```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## 设置模型
### 具体模型，dense layer，optimizer什么意思，在这里不用在意。
这行代码规定了一个模型，模型包含一个dense layer，整个layer只有一个单位。输入输出都是一维。

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=(1,))])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.summary()
```

## 准备数据

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=int)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=int)
```

## 开始训练

```python
model.fit(xs, ys, epochs=500)
```

## 应用结果

```python
print(model.predict([10]))
```

## 评价结果

```python
testx = np.random.randint(10000,size=1000)
testy = testx*2-1
tol = 0.5
yhat = model.predict(testx)
```

```python
print(np.sum(np.abs(yhat-testy.reshape(-1,1))<tol))
```

```python
tol = 0.1
print(np.sum(np.abs(yhat-testy.reshape(-1,1))<tol))
```
